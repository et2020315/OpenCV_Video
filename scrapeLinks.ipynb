```
def get_youtube_link_to_txt(htmlname,folder):
  
  filelink = []
  hfile = open(htmlname)
  soup = BeautifulSoup(hfile, 'html.parser')

  for link in soup.find_all("a",href=True):
    if link['href'].find("//youtube.com") != -1:
      filelink.append(link["href"])
      # print(link["href"])
  print(filelink)
  hfile.close()
  createfile = htmlname.replace("html","txt").replace("HTMLfile","TextLinkFiles").replace("/"+folder,"")
  print("createfile:"+createfile)
  txtwrite = open(createfile, "w+")
  for i in range(len(filelink)):
    print(filelink[i])
    txtwrite.write(filelink[i] + "\n")
  txtwrite.close()
  print("write sucessfully")
```
