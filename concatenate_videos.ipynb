```
# main function
def compVideos(pathspecify,topicName,frate):
  try:
    # set video output
    outputPath = "/content/drive/My Drive/TestVideos/ByTopic/"+topicName + ".avi"
    codec = cv2.VideoWriter_fourcc("X","V","I","D")
    outputVideo = cv2.VideoWriter(outputPath,codec,frate,(630,360))
    # input video
    videoslst = glob.glob(pathspecify)
    print(videoslst)
    for f in videoslst:

      cap = cv2.VideoCapture(f)
      if cap.isOpened():
        while(True):
          ret, img = cap.read()
          if ret == True:
            w,h,l = img.shape
            if (h!= 630 or w != 360):
              img = cv2.resize(img,(630,360))
              print("resize!")
              
            else:
              print(str(img.shape))
            outputVideo.write(img)
          else:
            print("ret false")
            break
      else:
        print("not opened")
      if cap.isOpened():
        cap.release()
        print("release cap")

      print("end of iter")
    print("finish writing")
    if outputVideo.isOpened():
        outputVideo.release()
        print("release output")
  except:
    traceback.print_exc()
    
```
